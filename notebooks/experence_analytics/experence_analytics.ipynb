```python
def n_frequency(df, field, order='desc', n=10):
    result_df = (
        df[field]
        .value_counts()
        .reset_index()
        .rename(columns={field: 'Frequency', 'index': field})
        .sort_values(by='Frequency', ascending=(order.lower() == 'desc'))
        .head(n)
    )
    return result_df
```

```python
def calculate_average_for_throughput(df):
    df_copy = df.copy()
    df_copy['Average_throughput'] = df_copy['Average_throughput'] / 2
    result_df = (
        df_copy.groupby('Handset Type')
        .agg(Avg=('Average_throughput', 'mean'))
        .reset_index()
        .sort_values(by='Avg', ascending=False)
    )
    return result_df
```

```python
## create new data frame that has MSISDN/Number	Average RTT	Average TCP	Average throughput
```

```python
kmean_df = pd.DataFrame(columns=['MSISDN/Number','Average_throughput','Average_TCP', 'Average_RTT' ])
```

```python
kmean_df.head()
```

```python
kmean_df['MSISDN/Number'] = new_df['MSISDN/Number']
kmean_df['Average_throughput'] = new_df['Average_throughput'] / 2
kmean_df['Average_TCP'] = new_df['Average_TCP'] / 2
kmean_df['Average_RTT'] = new_df['Average_RTT'] / 2
```

```python
kmean_df
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = MinMaxScaler()
min_max_scaler = MinMaxScaler()
```

```python
def handle_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    column = column.apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
    return column
```

```python

# Selecting columns for normalization
columns_to_normalize = ['Average_throughput', 'Average_TCP', 'Average_RTT']

# Handling outliers
for column in columns_to_normalize:
    kmean_df[column] = handle_outliers(kmean_df[column])
    
# Normalizing the DataFrame
scaler = MinMaxScaler()
normal_df = scaler.fit_transform(kmean_df[columns_to_normalize])
# Convert the NumPy array to a DataFrame
normal_df = pd.DataFrame(normal_df, columns=columns_to_normalize)
# Concatenating with the 'MSISDN/Number' column
new_normalized_df = pd.concat([kmean_df['MSISDN/Number'], normal_df], axis=1)
new_normalized_df
```

```python
from sklearn.cluster import KMeans
```

```python
features_for_clustering = ['Average_throughput', 'Average_TCP', 'Average_RTT']

# Extracting the features for clustering
X = new_normalized_df[features_for_clustering]
num_clusters = 3

kmeans = KMeans(n_clusters=num_clusters, random_state=42)
new_normalized_df['Cluster'] = kmeans.fit_predict(X)

# Description of each cluster
cluster_descriptions = []

for cluster_num in range(num_clusters):
    cluster_data = new_normalized_df[new_normalized_df['Cluster'] == cluster_num][features_for_clustering]
    
    # Calculate cluster statistics (mean, std, etc.) based on your understanding of the data
    
    cluster_mean = cluster_data.mean()

    description = f"Cluster {cluster_num + 1}: "
    description += f"Average_throughput: {cluster_mean['Average_throughput']:.2f}, "
    description += f"Average_TCP: {cluster_mean['Average_TCP']:.2f}, "
    description += f"Average_RTT: {cluster_mean['Average_RTT']:.2f}"
    
    cluster_descriptions.append(description)

# Print cluster descriptions
for desc in cluster_descriptions:
    print(desc)
```
