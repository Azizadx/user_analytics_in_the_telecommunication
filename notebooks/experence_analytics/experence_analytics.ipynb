#### Task 4.2 - Compute & list 10 of the top, bottom and most frequent:

  - [X] TCP values in the dataset. 
  - [X] RTT values in the dataset.
  - [X] Throughput values in the dataset.

```python
def n_frequency(df, field, order='desc', n=10):
    result_df = (
        df[field]
        .value_counts()
        .reset_index()
        .rename(columns={field: 'Frequency', 'index': field})
        .sort_values(by='Frequency', ascending=(order.lower() == 'desc'))
        .head(n)
    )
    return result_df
```

```python
field_name = 'TCP UL Retrans. Vol (Bytes)'
order_type = 'asc'
top_10_result = n_frequency(new_df, field_name, order_type)
top_10_result
```
